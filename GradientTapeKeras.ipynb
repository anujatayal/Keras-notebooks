# Gradient Tape

```python
x=tf.constant(5.0)
with tf.GradientTape() as tape:
    tape.watch(x)
    y=x**3
tape.gradient(y,x)
```

```python
x=tf.Variable(5.0,trainable=True)
with tf.GradientTape() as tape:
    y=x**2
tape.gradient(y,x)
```

# Higher Order Derivative

```python
x=tf.Variable(3.0,trainable=True)
with tf.GradientTape() as tape1:
    with tf.GradientTape() as tape2:
        y=x**3
    order1=tape2.gradient(y,x)
order2=tape1.gradient(order1,x)
print(order1)
print(order2)
```

# Multi Variable Derivative

```python
#without persistent error
x=tf.Variable(3.0,trainable=True)
y=tf.Variable(2.0,trainable=True)
with tf.GradientTape(persistent=True) as tape:
    y1=x**2
    y2=y**2
print(tape.gradient(y1,x))
print(tape.gradient(y2,y))
```

```python
import numpy as np
import random
```

```python
def loss(real,pred):
    return(tf.abs(real-pred))
xtrain=np.asarray(range(0,11))
ytrain=np.asarray([i*10+5 for i in xtrain])
a=tf.Variable(random.random(),trainable=True)
b=tf.Variable(random.random(),trainable=True)
```

```python
print(a)
print(b)
```

```python
def step(realX,realY):
    with tf.GradientTape(persistent=True) as tape:
        predY=a*realX+b
        regLoss=loss(realY,predY)
    aGrad,bGrad=tape.gradient(regLoss,(a,b))
    a.assign_sub(aGrad*0.001)
    b.assign_sub(bGrad*0.001)
```

```python
for i in range(10000):
    step(xtrain,ytrain)
    print(i,a.numpy(),b.numpy())
print(a.numpy(),b.numpy())
```
